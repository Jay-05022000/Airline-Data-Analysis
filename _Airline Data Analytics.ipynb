```python
# Import necessary libraries & create a dataframe.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Data Cleaning  : 
1) As we know that whole dataset is of the year 2020,so there is no need of having a 'year' column in it.
2) As all of our aims revolve around flight delay,we can drop the rows in which there is no flight delay.
3) Drop 'Flights' and 'Airtime'columns as they won't be needed in the analysis.

```python
df1=df1.drop(columns=['Year']) # Dropping the 'Year' column.

# As there are two types of delays: Departure delay and Arrival delay,rows in which values of both these columns are 0 are not usefull.Hence will remove it.

df1=df1.drop(df1[(df1['DepDelayMinutes'] == 0)  & (df1['ArrDelayMinutes'] == 0)].index) 
df1=df1.drop(columns=['AirTime','Flights']) # Dropping 'AirTime' and 'Flights' columns.
df1
```

```python
df1.to_excel('Cleaned Airline 2020 dataset.xlsx') # Exporting Cleaned dataset to a excel file.  
```

## 1) Which airports should be focused more on infrastructure development to mitigate the frequency and severity of delays? 

Observation:After reviewing the dataset, it seems that a flight may not encounter delays in arrival simply due to experiencing a delay in departure.Alternatively, not every flight that experiences a delay in departure also encounters delays in arrival.  

```python
Origin_Depart_Delay=df1[["Origin","DepDelayMinutes"]] # Creating a dataframe specifically to solve first question.

Origin_Depart_Delay.head() 
```

```python
 # Remove the rows which has DepDelayMinutes=0 and null values(because these flights are cancelled for some reasons).
    
Origin_Depart_Delay=Origin_Depart_Delay.drop(Origin_Depart_Delay[Origin_Depart_Delay['DepDelayMinutes'] == 0].index)
Origin_Depart_Delay=Origin_Depart_Delay.dropna()
Origin_Depart_Delay=Origin_Depart_Delay.reset_index(drop=True)
 
Origin_Depart_Delay
```

```python
Origin_Depart_Delay= Origin_Depart_Delay.groupby('Origin').size().sort_values(ascending=False) # Calculating the count of delayed departure flights at each airport.
Origin_Depart_Delay= Origin_Depart_Delay.to_frame() # Converting pandas series into pandas dataframe.

Origin_Depart_Delay=Origin_Depart_Delay.rename(columns={"Origin":"Departure Airport",0:"No of Delayed Departures"}) # Renaming the column names.
Origin_Depart_Delay=Origin_Depart_Delay.reset_index()
Origin_Depart_Delay
```

```python
Origin_Depart_Delay.head(5)
```

```python
# Creating a visualisation.

Origin_Depart_Delay.head(10).plot(kind='bar', x='Origin', y='No of Delayed Departures', figsize=(50 ,20))
plt.title('Number of Departure Delay Per Airport',fontsize=34)
plt.xlabel('Departure Airport',fontsize=34)
plt.ylabel('No of Delayed Departures',fontsize=34)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.grid()
plt.show()
```

Finding :The airports facing the highest departure delays are ATL (137), DFW (115), ORD (103), CLT (98), and DEN (93). Therefore, these airports require infrastructural modifications to address and alleviate the delays.

# 2) What are the relationships between airline carriers and flight delays/cancellations?

The dataset df1 includes flight information, where instances of either departure delays, arrival delays, or flight cancellations are present. Therefore, there is no necessity to analyze both 'DepDelayMinutes' and 'ArrDelayMinutes' simultaneously to establish the correlation between airlines and flight delays. 

```python
# The df1 dataset contains flight details in which there is either a departure delay,arrival delay or a flight is cancelled.Hence we don't have to work with both 'DepDelayMinutes' and 'ArrDelayMinutes' to draw the relationship between airlines and flight delays.

Airline_flight_Delays=df1[["Reporting_Airline","DepDelayMinutes"]].reset_index(drop=True)
Airline_flight_Delays=Airline_flight_Delays.dropna()  # Removing cancelled flights.
Airline_flight_Delays 
```

```python
Airline_flight_Delays=Airline_flight_Delays.groupby('Reporting_Airline').size().sort_values(ascending=False).to_frame()
Airline_flight_Delays=Airline_flight_Delays.reset_index()
Airline_flight_Delays=Airline_flight_Delays.rename(columns={"Reporting_Airline":"Reporting_Airline",0:"No of flight delays"})

Airline_flight_Delays  
```

```python
# Creating a visualisation.

data=Airline_flight_Delays['No of flight delays'] 
label=Airline_flight_Delays['Reporting_Airline'] 

plt.pie(data,labels=label,autopct='%1.2f%%',startangle=0,radius=1.55,pctdistance=0.75, labeldistance= 1.05)
plt.title('Percentage of flight delays by airlines',loc='center',pad= 70)
plt.show()
```

Finding : Following 5 Airlines account for 63% of the total flight delays recorded.

1) WN
2) AA
3) DL
4) OO
5) UA

# 3) What are the most common reasons for flight delays based on airports?

The ArrDelayMinutes is categorized into 5 types: CarrierDelay,WeatherDelay,NASDelay,SecurityDelay and LateAircraftDelay.
 

```python
Airport_Flight_delay=df1[['Dest','CarrierDelay','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay']]
Airport_Flight_delay=Airport_Flight_delay.dropna().reset_index(drop=True) # Droopping cancelled flights.

Airport_Flight_delay
```

```python
# Map values in all columns except 'Dest' for the ease of calculation.(0=0 and non-zero=1)

Airport_Flight_delay['CarrierDelay']= Airport_Flight_delay['CarrierDelay'].map(lambda x:0 if x else 1)
Airport_Flight_delay['WeatherDelay']= Airport_Flight_delay['WeatherDelay'].map(lambda x:0 if x else 1)
Airport_Flight_delay['NASDelay']= Airport_Flight_delay['NASDelay'].map(lambda x:0 if x else 1)
Airport_Flight_delay['LateAircraftDelay']= Airport_Flight_delay['LateAircraftDelay'].map(lambda x:0 if x else 1)
     
Airport_Flight_delay 
```

```python
Airport_Flight_delay['SecurityDelay']= Airport_Flight_delay['SecurityDelay'].map(lambda x:0 if x else 1)

Airport_Flight_delay
```

```python
Airport_Flight_delay=Airport_Flight_delay.groupby('Dest').sum() 
Airport_Flight_delay.max() 
```

```python
Airport_Flight_delay=Airport_Flight_delay.sort_values(by='NASDelay',ascending=False).reset_index()
Top=Airport_Flight_delay.head(5)
Top
```

```python
# Creating visualisation.

y1=np.array(Airport_Flight_delay['Dest'])
x1=np.array(Airport_Flight_delay['CarrierDelay'])
x2=np.array(Airport_Flight_delay['WeatherDelay'])
x3=np.array(Airport_Flight_delay['NASDelay'])
x4=np.array(Airport_Flight_delay['SecurityDelay'])
x5=np.array(Airport_Flight_delay['LateAircraftDelay'])

bar_width=0.15
ytick_spacing = 0.3 
plt.figure(figsize=(15,40))
 
plt.barh(np.arange(len(y1)) , x1 , height=bar_width,color='m',label='CarrierDelay')
plt.barh(np.arange(len(y1)) + bar_width , x2 , height=bar_width,color='orange',label='WeatherDelay')
plt.barh(np.arange(len(y1)) + 2*bar_width , x3 , height=bar_width,color='green',label='NASDelay')
plt.barh(np.arange(len(y1)) + 3*bar_width , x4 , height=bar_width,color='cyan',label='SecurityDelay')
plt.barh(np.arange(len(y1)) + 4*bar_width , x5 , height=bar_width,color='darkblue',label='LateAircraftDelay')

yticks_labels = [f"{dest}\n" for dest in y1]
plt.yticks(np.arange(len(y1)) + 2*bar_width,yticks_labels )   

plt.xlabel('Type of Delay')
plt.ylabel('Airport')
plt.title('Type of delay per airport')
plt.legend()

plt.show()
```

Finding:Based on the findings above, it is evident that WeatherDelay and SecurityDelay are common causes of flight delays across airports. 

# 4) What are the relationships between days in a week or times in a day when there are a lot of cancellations and delays?

```python
DOW_Cancellations=df1[['DayOfWeek','Cancelled']]
DOW_Cancellations=DOW_Cancellations.drop(DOW_Cancellations[DOW_Cancellations['Cancelled'] == 0].index).reset_index(drop=True)
DOW_Cancellations
```

```python
DOW_Cancellations=DOW_Cancellations.groupby('DayOfWeek').size().to_frame().reset_index()
DOW_Cancellations=DOW_Cancellations.rename(columns={0:'No of Cancellations'})
DOW_Cancellations
```

```python
# Plotting a graph.

plt.plot(np.array(DOW_Cancellations['DayOfWeek']),np.array(DOW_Cancellations['No of Cancellations']),color='red',marker='o',markerfacecolor='blue')
plt.xlabel('Day Of A Week')
plt.ylabel('No of Cancellations')
plt.title('Total No of flight cancellations on a given day of a week')
plt.grid()
plt.show()
```

Finding:The 4th and 5th days of the week experienced a higher number of cancellations compared to any other weekdays, while the 6th day had the fewest cancellations.

Now let's see, at what time of the day more number of departure flights were delayed?

```python
Depart_Time_Slot=df1[['DepTimeBlk','DepDelayMinutes']]
Depart_Time_Slot=Depart_Time_Slot.drop(Depart_Time_Slot[Depart_Time_Slot['DepDelayMinutes'] == 0].index).reset_index(drop=True)
Depart_Time_Slot=Depart_Time_Slot.dropna()

Depart_Time_Slot
```

```python
Depart_Time_Slot=Depart_Time_Slot.groupby('DepTimeBlk').size().to_frame().reset_index()
Depart_Time_Slot=Depart_Time_Slot.rename(columns={'DepTimeBlk':'Departure Time Slot',0:'No of departure flight delays'})
Depart_Time_Slot

```

```python
# Plotting a graph.
plt.rcParams["figure.figsize"] = (24, 10)

plt.plot(np.array(Depart_Time_Slot['Departure Time Slot']),np.array(Depart_Time_Slot['No of departure flight delays']),color='green',marker='s',markerfacecolor='yellow',markersize=8)
plt.xlabel('Departure Time Slot',fontsize=24)
plt.ylabel('No of departure flight delays',fontsize=24)
plt.title('Total No of departure delays per time slots',fontsize=24)
plt.grid()
plt.show()
```

Finding:Between 5:00 PM and 7:59 PM, an average of 175 departure flights were canceled per hour.

```python
Arrival_Delay_Time_Slot=df1[['ArrTimeBlk','ArrDelayMinutes']]
Arrival_Delay_Time_Slot=Arrival_Delay_Time_Slot.drop(Arrival_Delay_Time_Slot[Arrival_Delay_Time_Slot['ArrDelayMinutes'] == 0].index) # Removing records that doesn't have arrival delay.
Arrival_Delay_Time_Slot=Arrival_Delay_Time_Slot.dropna() # Removing null values
Arrival_Delay_Time_Slot 
```
